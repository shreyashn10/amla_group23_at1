### G.4 Model Technical Performance

> Provide some explanations on model performance


```python
# ------------------- Training Performance -------------------
y_train_pred = model.predict(X_train)
y_train_proba = model.predict_proba(X_train)[:, 1]
print("Training Accuracy:", accuracy_score(y_train, y_train_pred))

plot_confusion_matrix(y_train, y_train_pred, labels=model.classes_, title="Training Confusion Matrix")
plot_roc_auc(y_train, y_train_proba, title="Training ROC Curve")
plot_precision_recall_curve(y_train, y_train_proba, title="Training Precision-Recall Curve")
```

```python
# ------------------- Validation Performance -------------------
y_val_pred = model.predict(X_val)
y_val_proba = model.predict_proba(X_val)[:, 1]
print("Validation Accuracy:", accuracy_score(y_val, y_val_pred))

plot_confusion_matrix(y_val, y_val_pred, labels=model.classes_, title="Validation Confusion Matrix")
plot_roc_auc(y_val, y_val_proba, title="Validation ROC Curve")
plot_precision_recall_curve(y_val, y_val_proba, title="Validation Precision-Recall Curve")
```

```python
# <Student to fill this section>
model_performance_explanations = """
The Logistic Regression model achieved high accuracy on both the training (0.997) and validation (0.995) sets. However, given the significant class imbalance (as shown in the class distribution plot), accuracy is not a reliable metric. The confusion matrices reveal the model's limitations: while it correctly identifies most of the "not drafted" players (high true negatives), it struggles to correctly identify the "drafted" players (low true positives). The training confusion matrix shows 70 true positives and 24 false negatives, while the validation confusion matrix shows only 16 true positives and 8 false negatives. This indicates poor recall for the minority class. The ROC curves show an AUC of 1.00 for both training and validation, which is suspiciously high and might suggest data leakage or an overly optimistic evaluation due to the imbalance. The precision-recall curves further highlight the issue with the minority class, showing a steep drop in precision as recall increases, indicating that many of the predicted "drafted" players are actually "not drafted" (low precision). Overall, despite high accuracy and AUC, the model's ability to identify actual drafted players is poor due to the class imbalance.
"""
```
